```python
import PyPDF2

def extract_hebrew_text_from_pdf(pdf_path):
    """
    Extracts text from a given Hebrew PDF file. This function reads a PDF and 
    returns its extracted text, assuming the content is in Hebrew or Right-To-Left (RTL) language.

    Parameters:
    - pdf_path: str : Path to the PDF file.

    Returns:
    - str: Extracted Hebrew text from the PDF.
    """
    try:
        # Open the PDF file
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            extracted_text = ""

            # Iterate over each page and extract text
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                page_text = page.extract_text()

                if page_text:
                    # You can reverse the text if needed for RTL languages
                    # extracted_text += page_text[::-1] + '\n'
                    
                    extracted_text += page_text + '\n'

        return extracted_text

    except Exception as e:
        print(f"An error occurred: {e}")
        return ""

def display_extracted_hebrew_text(pdf_path):
    """
    Displays extracted Hebrew text from a given PDF.

    Parameters:
    - pdf_path: str : Path to the PDF file.
    """
    hebrew_text = extract_hebrew_text_from_pdf(pdf_path)
    
    if hebrew_text:
        print("Extracted Hebrew Text:")
        print(hebrew_text)
    else:
        print("No text extracted or an error occurred.")

# Example usage in Jupyter Notebook
# Path to your PDF file
pdf_path = 'clean_document_heb.pdf'  # Replace with the actual path to your Hebrew PDF file

# Call the function to display extracted Hebrew text
display_extracted_hebrew_text(pdf_path)
```
