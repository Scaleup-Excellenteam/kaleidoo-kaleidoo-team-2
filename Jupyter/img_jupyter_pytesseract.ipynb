```python
import pytesseract
from PIL import Image

def extract_text_from_image(image_path, lang='eng', tess_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'):
    """
    Extracts text from an image using Tesseract OCR.

    Parameters:
    - image_path: str : Path to the image file.
    - lang: str : Language code for the text extraction. Default is 'heb' for Hebrew.
    - tess_cmd: str : Path to the Tesseract-OCR executable.

    Returns:
    - str: Extracted text from the image.
    """
    try:
        # Set the path to the Tesseract executable
        pytesseract.pytesseract.tesseract_cmd = tess_cmd
        
        # Open the image file
        img = Image.open(image_path)
        
        # Set custom configuration for Tesseract
        custom_config = f'--oem 3 --psm 6 -l {lang}'
        
        # Perform OCR on the image
        extracted_text = pytesseract.image_to_string(img, config=custom_config)
        
        return extracted_text
    
    except Exception as e:
        print(f"An error occurred: {e}")
        return ""

def display_extracted_text(image_path, lang='eng'):
    """
    Displays extracted text from an image.

    Parameters:
    - image_path: str : Path to the image file.
    - lang: str : Language code for the text extraction. Default is 'heb' for Hebrew.
    """
    text = extract_text_from_image(image_path, lang=lang)
    
    if text:
        print("Extracted Text:")
        print(text)
    else:
        print("No text extracted or an error occurred.")

# Example usage
if __name__ == "__main__":
    # Path to your image file
    image_path = 'clean_text_image_eng.png'  
    
    # Display the extracted text
    display_extracted_text(image_path, lang='eng')
```
